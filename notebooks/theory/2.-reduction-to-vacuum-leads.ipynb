## Particle–Hole Transformation: Mapping to an Empty Initial State

### Motivation

The initial state has the **left lead and dot site $d_0$ fully occupied**, while the right lead
and dot site $d_1$ are empty. Numerically and analytically it is far more convenient to work
with a vacuum initial state. We achieve this by performing a **particle–hole (Bogoliubov)
transformation** on every mode that is occupied at $t=0$.

---

### The Transformation

Define new operators by swapping creation and annihilation on the occupied sector:

$$
\boxed{
\tilde{c}_{L,k} \;=\; c^{\dagger}_{L,k}, \qquad
\tilde{c}^{\dagger}_{L,k} \;=\; c_{L,k},
\qquad\qquad
\tilde{d}_0 \;=\; d^{\dagger}_0, \qquad
\tilde{d}^{\dagger}_0 \;=\; d_0.
}
$$

All other operators are left **unchanged**:

$$
\tilde{c}_{R,k} = c_{R,k}, \qquad \tilde{d}_1 = d_1.
$$

One can verify directly that $\{\tilde{c}_{L,k},\tilde{c}^{\dagger}_{L,k'}\}=\delta_{kk'}$
and $\{\tilde{d}_0,\tilde{d}^{\dagger}_0\}=1$, so the canonical anti-commutation
relations are preserved — this is a valid Bogoliubov transformation.

#### New vacuum

The transformed vacuum $|\tilde 0\rangle$ is defined by

$$
\tilde{c}_{L,k}\,|\tilde 0\rangle = 0,\quad
\tilde{d}_0\,|\tilde 0\rangle = 0,\quad
c_{R,k}\,|\tilde 0\rangle = 0,\quad
d_1\,|\tilde 0\rangle = 0 \qquad \forall\, k.
$$

In terms of the original operators this state is precisely the physical initial state:
every left-lead mode and $d_0$ are occupied, while the right lead and $d_1$ are empty.

---

### Transformed Hamiltonian

We now express $H$ in terms of the tilde operators. Each piece is worked out in turn.

#### Left lead — band inversion

$$
H_L = -t\sum_k\!\left(c^{\dagger}_{L,k+1}c_{L,k}+\mathrm{h.c.}\right)
\;\xrightarrow{\;\text{p-h}\;}\;
+t\sum_k\!\left(\tilde{c}^{\dagger}_{L,k+1}\tilde{c}_{L,k}+\mathrm{h.c.}\right).
$$

The hopping amplitude **changes sign**: holes hop with amplitude $+t$, corresponding to
a band inversion $\varepsilon_k \to -\varepsilon_k$.  No constant is generated because
off-diagonal hoppings anticommute without producing a $\delta_{kk'}$ term.

#### Right lead — unchanged

$$
H_R = -t\sum_k\!\left(c^{\dagger}_{R,k+1}c_{R,k}+\mathrm{h.c.}\right) \quad \text{(no change)}.
$$

#### Dot single-particle energies

Using $\hat n_0 = d^\dagger_0 d_0 = \tilde d_0\tilde d^\dagger_0 = 1 - \tilde d^\dagger_0\tilde d_0$:

$$
H_{\text{dot}} = \varepsilon_d(\hat n_0 + \hat n_1)
\;\longrightarrow\;
-\varepsilon_d\,\tilde{\hat n}_0 \;+\; \varepsilon_d\,\hat n_1 \;+\; \varepsilon_d.
$$

The on-site energy of the hole on $\tilde d_0$ is $-\varepsilon_d$
(the gate voltage flips sign for holes).

#### Coulomb interaction

Using $\hat n_0 = 1 - \tilde{\hat n}_0$:

$$
H_{\text{int}} = U\hat n_0\hat n_1
= U\bigl(1 - \tilde{\hat n}_0\bigr)\hat n_1
= U\hat n_1 - U\,\tilde{\hat n}_0\,\hat n_1.
$$

This generates a **constant shift $+U$ of the $d_1$ level**: because $d_0$ was always
occupied before the quench, $d_1$ always felt the Coulomb energy $U$ from it.
Combining with the $H_{\text{dot}}$ contribution, the effective on-site energy of $d_1$
becomes

$$
\varepsilon_d + U \;\equiv\; \tilde\varepsilon_1.
$$

#### Lead–dot coupling

$$
-t\!\left(c^{\dagger}_{L,0}\,d_0+\mathrm{h.c.}\right)
\;\longrightarrow\;
-t\!\left(\tilde c_{L,0}\,\tilde d^\dagger_0+\mathrm{h.c.}\right)
= +t\!\left(\tilde c^{\dagger}_{L,0}\,\tilde d_0+\mathrm{h.c.}\right).
$$

The sign of the left coupling flips (equivalent to a phase redefinition; physically
irrelevant). The right coupling is unaffected:

$$
-t\!\left(c^{\dagger}_{R,0}\,d_1+\mathrm{h.c.}\right)\quad\text{(unchanged)}.
$$

---

### Full Transformed Hamiltonian

Collecting all terms and dropping the irrelevant constant $\varepsilon_d$:

$$
\boxed{
\begin{aligned}
\tilde H &=
+t\!\sum_k\!\left(\tilde c^{\dagger}_{L,k+1}\tilde c_{L,k}+\mathrm{h.c.}\right)
-t\!\sum_k\!\left(c^{\dagger}_{R,k+1}c_{R,k}+\mathrm{h.c.}\right)
-\varepsilon_d\,\tilde{\hat n}_0
+(\varepsilon_d+U)\hat n_1
-U\,\tilde{\hat n}_0\,\hat n_1
\\
&\quad
+t\!\left(\tilde c^{\dagger}_{L,0}\,\tilde d_0+\mathrm{h.c.}\right)
-t\!\left(c^{\dagger}_{R,0}\,d_1+\mathrm{h.c.}\right)
\end{aligned}
}
$$

with **vacuum initial condition** $|\tilde\Psi(0)\rangle = |\tilde 0\rangle$.

---

### Summary of Changes

| Quantity | Original | After transformation |
|---|---|---|
| Left-lead hopping | $-t$ | $+t$ (band inversion) |
| Right-lead hopping | $-t$ | $-t$ (unchanged) |
| $d_0$ on-site energy | $+\varepsilon_d$ | $-\varepsilon_d$ (hole level) |
| $d_1$ on-site energy | $+\varepsilon_d$ | $+\varepsilon_d + U$ (Coulomb shift) |
| Interaction $U\hat n_0\hat n_1$ | $+U$ | $-U$ (sign flip) |
| Left coupling | $-t$ | $+t$ (sign flip) |
| Right coupling | $-t$ | $-t$ (unchanged) |
| Initial state | filled left + $d_0$, empty right + $d_1$ | **all empty** $|\tilde 0\rangle$ |

The physics is unchanged — only the reference state has been shifted from a
complicated filled Fermi sea to the convenient empty vacuum.
