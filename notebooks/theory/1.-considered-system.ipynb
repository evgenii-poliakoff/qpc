# Coulomb Blockade in a Two-Site Quantum Dot: Lecture Notes

---

## 1. The Hamiltonian

The full Hamiltonian of the system is:

$$H = H_L + H_R + H_{\text{dot}} + H_{\text{int}} + H_{\text{coup}}$$

Each piece is described below.

---

### 1.1 Left Lead

$$H_L = -t \sum_{k=0}^{\infty} \left( c^{\dagger}_{L,k+1} \, c_{L,k} + c^{\dagger}_{L,k} \, c_{L,k+1} \right)$$

- $c^{\dagger}_{L,k}$, $c_{L,k}$: creation/annihilation operators for spinless electrons on site $k$ of the left lead.
- $t > 0$: nearest-neighbour hopping amplitude (same for both leads).
- The lead is a **semi-infinite tight-binding chain**, with a cosine dispersion band $\varepsilon(k) \in [-2t, +2t]$.

---

### 1.2 Right Lead

$$H_R = -t \sum_{k=0}^{\infty} \left( c^{\dagger}_{R,k+1} \, c_{R,k} + c^{\dagger}_{R,k} \, c_{R,k+1} \right)$$

Identical in structure to $H_L$. Both leads are non-interacting and spinless.

---

### 1.3 Central Dot (Single-Particle Energies)

$$H_{\text{dot}} = \varepsilon_d \left( d^{\dagger}_0 d_0 + d^{\dagger}_1 d_1 \right)$$

- $d^{\dagger}_0$, $d^{\dagger}_1$: creation operators for spinless electrons on dot sites $0$ and $1$.
- $\varepsilon_d$: **gate voltage** — the tunable on-site energy of each dot level. This is the key control parameter for Coulomb blockade oscillations.

---

### 1.4 Coulomb Interaction

$$H_{\text{int}} = U \, d^{\dagger}_0 d_0 \, d^{\dagger}_1 d_1 \equiv U \, \hat{n}_0 \, \hat{n}_1$$

- $U > 0$: Coulomb repulsion energy, active **only** when both dots are simultaneously occupied (double occupancy).
- $\hat{n}_i = d^{\dagger}_i d_i$ is the number operator for dot $i \in \{0, 1\}$.

---

### 1.5 Lead–Dot Coupling

$$H_{\text{coup}} = -t \left( c^{\dagger}_{L,0} \, d_0 + d^{\dagger}_0 \, c_{L,0} \right) - t \left( c^{\dagger}_{R,0} \, d_1 + d^{\dagger}_1 \, c_{R,0} \right)$$

- The **left lead** couples to dot $d_0$ via hopping amplitude $-t$.
- The **right lead** couples to dot $d_1$ via hopping amplitude $-t$.
- For simplicity the dot-lead hopping is taken equal to the bulk lead hopping $t$.

---

### 1.6 Full Hamiltonian (Collected)

$$\boxed{
H = -t\!\sum_{k}\!\left(c^{\dagger}_{L,k+1}c_{L,k} + \text{h.c.}\right)
-t\!\sum_{k}\!\left(c^{\dagger}_{R,k+1}c_{R,k} + \text{h.c.}\right)
+ \varepsilon_d(\hat{n}_0+\hat{n}_1)
+ U\hat{n}_0\hat{n}_1
- t\!\left(c^{\dagger}_{L,0}d_0+\text{h.c.}\right)
- t\!\left(c^{\dagger}_{R,0}d_1+\text{h.c.}\right)
}$$

---

## 2. Initial Conditions: The Pre-Quench State

Before the coupling $H_{\text{coup}}$ is switched on, the leads and dots are **decoupled**. The system is prepared in a specific initial state, described below.

---

### 2.1 Left Lead: Completely Filled

The left lead is initialised in its **ground state at zero temperature with all states filled**:

$$|\Psi_L\rangle = \prod_{k=0}^{\infty} c^{\dagger}_{L,k} |0\rangle$$

In terms of the band structure: the Fermi level of the left lead sits at the **top of the band**, $\mu_L = +2t$.

This is equivalent to saying the left reservoir is at **infinite chemical potential** (or zero temperature with a fully filled band). Every single-particle state in the left lead is occupied.

---

### 2.2 Right Lead: Completely Empty

The right lead is initialised in the **vacuum state**:

$$|\Psi_R\rangle = |0\rangle$$

The Fermi level of the right lead sits at the **bottom of the band** (or below it), $\mu_R = -2t$.

Every single-particle state in the right lead is unoccupied.

---

### 2.3 Central Dots: Initial Occupation

Before the quench, the dots are also decoupled from the leads. Their initial occupation is set by $\varepsilon_d$ alone (since $H_\text{coup} = 0$). In the simplest quench protocol:

- The dots start **empty**: $|\Psi_{\text{dot}}\rangle = |00\rangle$, or
- The dot occupation is irrelevant if $H_\text{coup}$ is switched on suddenly — the transient dynamics washes out the initial dot state quickly.

---

### 2.4 The Quench Protocol

At time $t = 0$, the coupling $H_{\text{coup}}$ is **suddenly switched on** (a quantum quench):

$$H(t) = \begin{cases} H_L + H_R + H_{\text{dot}} + H_{\text{int}} & t < 0 \\ H_L + H_R + H_{\text{dot}} + H_{\text{int}} + H_{\text{coup}} & t \geq 0 \end{cases}$$

The enormous chemical potential bias $\Delta\mu = \mu_L - \mu_R = 4t$ drives electrons from the left lead through the dot system into the right lead. This is the non-equilibrium current we study.

---

### 2.5 Effective Bias Window

Because the left lead is **full** and the right lead is **empty**, the effective bias window for transport covers the **entire bandwidth**:

$$\text{Bias window: } \mu_R = -2t \leq \varepsilon \leq \mu_L = +2t$$

Any many-body transition of the dot whose energy lies within this window is **accessible** and contributes to transport. This is a much wider window than in a typical small-bias experiment, but the physics of Coulomb blockade (which transitions are resonant) remains the same.

---

## 3. Why Two Peaks Are Observed as $\varepsilon_d$ Is Varied

This is the heart of Coulomb blockade physics. We now explain, step by step, why sweeping the gate voltage $\varepsilon_d$ produces **exactly two peaks** in the current.

---

### 3.1 Many-Body States of the Isolated Dot

The dot subspace has four basis states (spinless fermions on two sites):

| State | Notation | $N$ (electrons) | Energy |
|-------|----------|-----------------|--------|
| Both empty | $\|00\rangle$ | 0 | $E_0 = 0$ |
| Only dot 0 filled | $\|10\rangle$ | 1 | $E_1 = \varepsilon_d$ |
| Only dot 1 filled | $\|01\rangle$ | 1 | $E_1 = \varepsilon_d$ |
| Both filled | $\|11\rangle$ | 2 | $E_2 = 2\varepsilon_d + U$ |

The two $N=1$ states are **degenerate** (same energy $\varepsilon_d$) because both dots have the same on-site energy.

---

### 3.2 Addition Energies (Electrochemical Potentials)

The key quantities controlling transport are the **addition energies**: the energy cost to add the $N$-th electron to the dot.

$$\mu_1 \equiv E(N{=}1) - E(N{=}0) = \varepsilon_d$$

$$\mu_2 \equiv E(N{=}2) - E(N{=}1) = (2\varepsilon_d + U) - \varepsilon_d = \varepsilon_d + U$$

These are the **electrochemical potentials** of the dot for the $0\to1$ and $1\to2$ transitions, respectively. Note:

$$\mu_2 - \mu_1 = U$$

The two addition energies are always separated by exactly $U$, regardless of $\varepsilon_d$. Sweeping $\varepsilon_d$ **shifts both levels rigidly** up and down together.

---

### 3.3 The Transport Condition (Rate Equation Picture)

In the sequential tunnelling (weak coupling) picture, current can flow through the dot via a transition $N \to N \pm 1$ **only if** the corresponding addition energy lies inside the bias window:

$$\mu_R \leq \mu_N \leq \mu_L$$

If $\mu_N$ is outside the window, that charge transition is **energetically forbidden** and the dot is trapped (blocked) in charge state $N$.

---

### 3.4 Three Regimes as $\varepsilon_d$ Is Swept

Let us trace what happens as $\varepsilon_d$ is swept from large positive to large negative values. The bias window is $[-2t, +2t]$.

**Regime A: $\varepsilon_d \gg 2t$ (both levels above the window)**

$$\mu_1 = \varepsilon_d > 2t, \qquad \mu_2 = \varepsilon_d + U > 2t$$

Adding even one electron to the dot costs more energy than the leads can provide. The dot stays in $N=0$ (empty). **No current flows.**

---

**Regime B: Peak 1 — $\mu_1 \approx 0$ (inside window)**

As $\varepsilon_d$ decreases and passes through the bias window, $\mu_1 = \varepsilon_d$ enters the window:

$$-2t \leq \varepsilon_d \leq +2t$$

The $N=0 \leftrightarrow N=1$ transition is now resonant. An electron can hop: Left lead $\to$ dot $\to$ Right lead. **Current flows — this is Peak 1.**

At this point $\mu_2 = \varepsilon_d + U \approx U$, which is above the window (assuming $U > 4t$). So double occupancy is still blocked. The dot oscillates between $N=0$ and $N=1$.

---

**Regime C: Coulomb Blockade Valley**

After $\varepsilon_d$ drops below $-2t$, $\mu_1$ exits the bottom of the window. Now:

$$\mu_1 = \varepsilon_d < -2t \quad \Rightarrow \quad N=0 \to N=1 \text{ transition blocked from above}$$
$$\mu_2 = \varepsilon_d + U \quad \text{still above } +2t \text{ if } U \text{ is large enough}$$

The dot is **stuck in $N=1$**: it cannot go back to $N=0$ (costs energy to remove the electron) nor proceed to $N=2$ (costs too much energy to add a second electron). **No net current flows — this is the Coulomb blockade valley.**

The width of this valley in gate voltage is:

$$\Delta\varepsilon_d^{\text{valley}} \approx U - 4t$$

(the gap between $\mu_2$ entering the window and $\mu_1$ leaving it, accounting for the finite bandwidth $4t$). For $U \gg t$ this width approaches $U$.

---

**Regime D: Peak 2 — $\mu_2 \approx 0$ (inside window)**

Continuing to decrease $\varepsilon_d$, eventually $\mu_2 = \varepsilon_d + U$ enters the bias window:

$$-2t \leq \varepsilon_d + U \leq +2t \quad \Rightarrow \quad -2t - U \leq \varepsilon_d \leq +2t - U$$

Now the $N=1 \leftrightarrow N=2$ transition is resonant. An electron can hop: Left lead $\to$ dot (now doubly occupied) $\to$ Right lead. **Current flows — this is Peak 2.**

---

**Regime E: $\varepsilon_d \ll -U - 2t$ (both levels below the window)**

Both addition energies are now below $-2t$:

$$\mu_1 = \varepsilon_d < -2t, \qquad \mu_2 = \varepsilon_d + U < -2t$$

The dot fills completely to $N=2$ and is stuck there. No transitions are accessible. **No current flows.**

---

### 3.5 Summary Diagram

```
Energy
  ^
  |
+2t ----[===== Bias Window =====]----
  |
  |     Regime A    |  Peak 1  | Blockade | Peak 2 |  Regime E
  |     (N=0 stuck) |  (N=0↔1) | (N=1)    | (N=1↔2)| (N=2 stuck)
  |
-2t ----[===== Bias Window =====]----
  |
  +-----> ε_d decreasing (gate voltage swept negative)
```

As a function of $\varepsilon_d$, the current profile looks like:

```
Current
  ^
  |        *                       *
  |       * *                     * *
  |      *   *                   *   *
  |_____*     *_________________*     *_____
  +-----------------------------------------> ε_d
      Peak 1      Blockade Valley      Peak 2
                  (width ≈ U)
```

The two peaks are separated in gate voltage by:

$$\boxed{\Delta\varepsilon_d = U}$$

---

### 3.6 Physical Intuition

The two peaks can be understood with a simple electrostatic picture:

- **Peak 1** is the resonance of the *first* electron entering the dot. The Fermi level of the lead aligns with the bare dot level $\varepsilon_d$.
- **Peak 2** is the resonance of the *second* electron. To enter, it must overcome the Coulomb repulsion of the first electron already sitting on the dot, so it only becomes resonant when $\varepsilon_d$ has been lowered by an additional $U$.
- In between, the dot is in a **definite charge state** $N=1$: removing the electron costs energy (disfavoured) and adding one costs energy due to $U$ (disfavoured). Transport is blocked.

This is entirely analogous to the **charging energy** of a classical capacitor: adding the second electron to a charged island requires extra energy $U = e^2/C$.

---

### 3.7 Effect of Finite Temperature and Linewidth

In a real system (or a numerical simulation with finite coupling $t$):

- **Finite coupling** broadens each peak by a tunnelling width $\Gamma \sim t^2 / \text{(lead DOS)}$. Peaks have a Lorentzian-like lineshape.
- **Finite temperature** $T$ smears the Fermi function, broadening peaks by $\sim k_B T$.
- The Coulomb blockade valley is visible as long as $U \gg \Gamma, k_B T$.
- In the quench setup, the current at long times reflects a **time-averaged** version of this physics, with the steady-state current playing the role of conductance.

---

## Quick Reference: Key Formulas

| Quantity | Formula |
|----------|---------|
| First addition energy | $\mu_1 = \varepsilon_d$ |
| Second addition energy | $\mu_2 = \varepsilon_d + U$ |
| Peak separation in gate voltage | $\Delta\varepsilon_d = U$ |
| Peak 1 position | $\varepsilon_d \approx \mu_{\text{lead}}$ |
| Peak 2 position | $\varepsilon_d \approx \mu_{\text{lead}} - U$ |
| Blockade valley width | $\approx U$ (for $U \gg t$) |
| Bias window (quench setup) | $\mu_R = -2t$, $\mu_L = +2t$ |


## 4. Coulomb Suppression of Current at $\varepsilon_d = 0$

### 4.1 Setup

We now fix the gate voltage at $\varepsilon_d = 0$ and ask: what happens to the steady-state
current as $U$ is increased from zero?

With $\varepsilon_d = 0$, the two addition energies are:

$$\mu_1 = \varepsilon_d = 0$$

$$\mu_2 = \varepsilon_d + U = U$$

The bias window is $[-2t, +2t]$, so $\mu_1 = 0$ sits exactly at the **centre** of the
window for all values of $U$. The first transition is always resonant.

---

### 4.2 The $U = 0$ Reference: Maximum Current

When $U = 0$, the interaction is absent and the two dots form a simple non-interacting
bridge between the leads. Both addition energies coincide: $\mu_1 = \mu_2 = 0$.

Both the $N{=}0 \leftrightarrow N{=}1$ and $N{=}1 \leftrightarrow N{=}2$ transitions are
simultaneously resonant at the centre of the bias window. The dot is free to be singly
or doubly occupied, and electrons traverse the junction without any additional energy
penalty. This corresponds to the **maximum current** at this gate voltage.

---

### 4.3 Finite $U$: Splitting the Two Transitions

As $U$ is turned on, the second addition energy $\mu_2 = U$ is pushed upward:

| $U$ | $\mu_1$ | $\mu_2$ | $\mu_2$ inside window $[-2t,+2t]$? |
|-----|---------|---------|--------------------------------------|
| $0$ | $0$ | $0$ | Yes (degenerate with $\mu_1$) |
| $t$ | $0$ | $t$ | Yes |
| $2t$ | $0$ | $2t$ | Yes (just at the edge) |
| $4t$ | $0$ | $4t$ | **No** — above $+2t$ |
| $\gg t$ | $0$ | $\gg 2t$ | No — deep in blockade |

There are two distinct physical regimes, separated by the crossover $U \approx 4t$.

---

### 4.4 Regime I — Partial Suppression: $0 < U \lesssim 4t$

Both transitions remain inside the bias window, but they are no longer degenerate.
The dot must spend time in the $N=1$ state between the two sequential tunnelling
events $0 \to 1 \to 2$. This slows down the current relative to the $U=0$ case for
two reasons:

1. **Asymmetry of the two resonances.** At $\varepsilon_d=0$ the first transition is
   perfectly centred, but the second transition at $\mu_2 = U$ sits off-centre. The
   tunnelling rate through an off-resonance level is reduced, creating a
   **bottleneck** for the $N{=}1 \to N{=}2$ step.

2. **Dwell time in the blocked intermediate state.** Because $\mu_2 > \mu_1$, the dot
   lingers in the $N=1$ charge state waiting for a fluctuation that can supply the
   extra energy $U$. This increases the average time between successive electron
   transfers and reduces the time-averaged current.

The current therefore **decreases monotonically** as $U$ grows from $0$ to $4t$.

---

### 4.5 Regime II — Full Coulomb Blockade: $U \gtrsim 4t$

Once $\mu_2 = U > 2t$, the second addition energy exits the top of the bias window.
The $N{=}1 \to N{=}2$ transition is now energetically **forbidden**: no state in either
lead can absorb the energy required to doubly occupy the dot.

The transport cycle that was possible for small $U$:

$$\text{Left lead} \xrightarrow{\mu_1 \in \text{window}} N{=}1
  \xrightarrow{\mu_2 \in \text{window}} N{=}2
  \xrightarrow{\text{empty to right}} \cdots$$

is now broken. Once the dot reaches $N=1$, it is **stuck**: removing the electron is
also suppressed because $\mu_1 = 0$ is at the centre of the window — the left lead
pushes electrons in as readily as the right lead accepts them — so there is no
preferred direction to empty the dot once it is singly occupied.

More precisely, $\mu_1 = 0$ being inside the window means the $N{=}0 \leftrightarrow
N{=}1$ transition is still active, but the **asymmetry** of the leads (left full, right
empty) means the dot is driven toward $N=1$ and held there. With $\mu_2$ above the
window, $N=2$ is inaccessible, and the right lead cannot accept a second sequential
electron. The current saturates to a **strongly suppressed** value that decays as $1/U$
for large $U$.

---

### 4.6 Summary: Current vs $U$ at Fixed $\varepsilon_d = 0$
```
Current
  ^
  |* (U=0, both transitions resonant)
  |
  | \
  |  \       (partial suppression,
  |   \       μ₂ drifting toward band edge)
  |    \
  |     '----.__
  |             '------._________  (full blockade plateau, ~1/U tail)
  +-------------------------------------> U / t
       0    2t   4t    6t    8t
              ^
              U ≈ 4t: μ₂ exits window
```

The crossover at $U \approx 4t$ marks the onset of **full Coulomb blockade**. For
$U \gg 4t$ the current is negligible and the dot is frozen in the $N=1$ state.

---

### 4.7 Why $\varepsilon_d = 0$ Is Special

The choice $\varepsilon_d = 0$ places the system at the **worst possible gate voltage**
for observing a clean blockade onset, because:

- $\mu_1 = 0$ is permanently resonant, so there is always a partial conduction channel.
- The suppression is **gradual** rather than sharp, reflecting the drift of $\mu_2$
  rather than a discrete level entering or leaving the window.

By contrast, tuning $\varepsilon_d = -U/2$ (the particle-hole symmetric point) keeps
the dot centred between the two transitions and produces the **sharpest** suppression
of current with increasing $U$, since the dot is maximally deep in the $N=1$ blockade
valley for all $U > 0$.
